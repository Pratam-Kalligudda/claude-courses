````python
from dotenv import load_dotenv
import os
load_dotenv()

if os.getenv("ANTHROPIC_API_KEY") is None: 
    raise Exception("api key not found")
````

````python
from anthropic import Anthropic

client = Anthropic()
model= "claude-haiku-4-5-20251001"
````

````python
def add_user_message(messages, user_prompt):
    messages.append({"role" : "user", "content" :user_prompt})

def add_assistant_message(messages, assistant_prompt):
    messages.append({"role" : "assistant", "content" :assistant_prompt})

def chat(messages, system=None, stop_tokens = None,temperature=0.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if system:
        params["system"] = system
    if stop_tokens:
        params["stop_sequences"] = stop_tokens
    
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
messages = []

prompt = """
   Generate three different AWS commands. Each should be very short.
"""

add_user_message(messages,prompt)
add_assistant_message(messages,"Here are the three AWS commands without any comments in a single bash block :\n```bash")
text = chat(messages, stop_tokens=["```"], temperature=1.0)
text.splitlines()
````

````python
import json

clean_json = json.loads(text.strip())
clean_json
````
