```python
from dotenv import load_dotenv
import os
load_dotenv()

if os.getenv("ANTHROPIC_API_KEY") is None: 
    raise Exception("api key not found")
```

```python
from anthropic import Anthropic

client = Anthropic()
model= "claude-haiku-4-5-20251001"
```

```python
def add_user_message(messages, user_prompt):
    messages.append({"role" : "user", "content" :user_prompt})

def add_assistant_message(messages, assistant_prompt):
    messages.append({"role" : "assistant", "content" :assistant_prompt})

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []
add_user_message(messages, "what is the meaning of life?")
assistant_response = chat(messages, "you should always answer in single sentence")
add_assistant_message(messages,assistant_response)
print(messages)
```
